Matplotlib:
It is useful for making plots

```python
# Importing Matplotlib Library
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

Plotting the data

```python
# sine wawe
plt.plot(x,y)
plt.show()
```

```python
# cosine wawe
plt.plot(x,z)
plt.show()
```

```python
# Adding title to X & Y axis

plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title("Sine wawe")
plt.show()
```

```python
# parabola

x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y, 'r+')
plt.show()
```

```python
plt.plot(x,y, 'g.')
plt.show()
```

```python
plt.plot(x,y, 'bx')
plt.show()
```

```python
x = np.linspace(-5,5, 50)
plt.plot(x, np.sin(x), 'g')
plt.plot(x, np.cos(x), 'r--')
plt.show()
```

Bar Plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
# 0,0 represents coordinates of origin
# 1,1 represents the height and weight of rectangle
# In simple terms it's just to mention a area in which we want our plot

languages = ["English", "French", "Spanish", "Latin", "German"]
people = [100,50,150,40,70]
ax.bar(languages, people)
plt.xlabel("Languages")
plt.ylabel("Number of People")
plt.show()
```

Pie Chart

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ["English", "French", "Spanish", "Latin", "German"]
people = [100,50,150,40,70]
ax.pie(people, labels=languages, autopct="%1.1f%%")
plt.show()
```

Scatter plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color="g")
ax.scatter(x,z,color="b")
plt.show()

# It is very useful in clustering applications
```

3D Scatter Plot

```python
fig3 = plt.figure()
ax = plt.axes(projection="3d")
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z, cmap="Greens")
plt.show()
```
